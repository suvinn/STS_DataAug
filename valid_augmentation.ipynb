## AEDA

```python
import random

def aeda(sentence):
    punc_list = list(".,;:?!")
    
    sentence = sentence.split()
    random_ratio = random.uniform(0.1, 0.3) # 범위는 ADEA 논문을 따름.
    n_ri = max(1, int(len(sentence) * random_ratio))
    
    for _ in range(n_ri):
        random_punc = random.choice(punc_list)
        random_idx = random.randint(0, len(sentence)-1)
        sentence.insert(random_idx, random_punc)
        
    return ' '.join(sentence).strip()
```

```python
rrest['sentence_1_bert'] = rrest['sentence_1'].apply(lambda x: aeda(x))
rrest['sentence_2_bert'] = rrest['sentence_2'].apply(lambda x: aeda(x))
rrest = rrest.rename(columns={
    'sentence_1_bert': 'sentence_1_aeda',
    'sentence_2_bert': 'sentence_2_aeda'
    })

rrest
```

```python
# (sen1_mod, sen2_mod)
both_mod = pd.DataFrame({
    'sentence_1': rrest['sentence_1_aeda'],
    'sentence_2': rrest['sentence_2_aeda']
})

both_mod
```

```python
# (sen1_mod, sen2)
sen1_ = rrest.drop(columns=['sentence_1', 'sentence_2_aeda'])
sen1_ = sen1_.rename(columns={
    'sentence_1_aeda': 'sentence_1'})
sen1_
```

```python
# (sen1, sen2_mod)
sen2_ = rrest.drop(columns=['sentence_2', 'sentence_1_aeda'])
sen2_ = sen2_.rename(columns={
    'sentence_2_aeda': 'sentence_2'})
sen2_
```

concat

```python
dataset = pd.read_csv('spacing_added.csv')
```

```python
sen12_ = pd.concat([sen1_, sen2_])
both = pd.concat([sen12_, both_mod])
both['index'] = both.index
both = both.sort_values(by='index')
both
```

```python
dataset = pd.concat([dataset, both])
dataset = dataset.sort_values(by='index')
dataset
```

```python
dataset = dataset.drop(columns='Unnamed: 0')
dataset
```

```python
dataset = pd.DataFrame(dataset)
dataset.to_csv('aeda_added.csv')
dataset
```
